```python
import matplotlib.pyplot as pl
pl.style.use('seaborn-whitegrid')
import numpy as np
```

```python
x=np.linspace(0,10,30)
y=np.sin(x)
pl.plot(x,y,'o',color='black')
```

```python
pl.scatter(x,y,color='black')
```

```python
rng = np.random.RandomState(2)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    pl.plot(rng.rand(2), rng.rand(2), marker,
    label="marker='{0}'".format(marker))
```

```python
pl.plot(x,y,'-ok')
# here - is line o is marker and k is last letter of black
```

```python
pl.scatter(x,y,marker='*',color='r',s=100)
```

```python
rng=np.random.RandomState(0)
a=rng.randint(1,100,(4,5))
print(a)
np.random.seed(0)
np.random.randint(1,100,(4,5))
# Both are same
```

```python
rng=np.random.RandomState(0)
x=rng.randn(100)
y=rng.randn(100)
colors=rng.rand(100)
sizes=1000*rng.rand(100)
pl.scatter(x,y,c=colors,s=sizes,alpha=0.3,cmap='viridis')
pl.colorbar()
```

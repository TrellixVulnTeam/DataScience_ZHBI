```python
import numpy as np
import matplotlib.pyplot as pl
```

```python
xlist=np.linspace(-3,3,3)
ylist=np.linspace(-3,3,4)
X,Y=np.meshgrid(xlist,ylist)
X,Y
```

```python
n,m=7,7
start=-3
x_vals=np.arange(start,start+n,1)
y_vals=np.arange(start,start+m,1)
X,Y=np.meshgrid(x_vals,y_vals)
X,Y
```

```python
fig,ax=pl.subplots()
ax.scatter(X,Y,color="green")
ax.set_title('Regular grid')
ax.set_xlabel('x')
ax.set_ylabel('y')
pl.show()
```

```python
xlist=np.linspace(-10.0,10.0,3)
ylist=np.linspace(-10.0,10.0,4)
X,Y=np.meshgrid(xlist,ylist)
Z=np.sqrt(X**2+Y**2)
pl.plot(Z)
X,Y,Z
```

```python
fig=pl.figure(figsize=(8,7))
ax=fig.add_axes([0.1,0.1,0.8,0.8])
cp=ax.contour(X,Y,Z)
ax.clabel(cp,inline=True)
ax.set_xlabel('x (cm)')
ax.set_ylabel('y (cm)')
pl.show()
```

```python
pl.figure(figsize=(8,7))
cp=pl.contour(X,Y,Z,colors='black',linestyles='dashed')
# we can also use linestyles=':','-','--' etc
pl.clabel(cp,inline=True,fontsize=10)
pl.xlabel('x (cm)')
pl.ylabel('y (cm)')
pl.show()
```

```python
fig = pl.figure(figsize=(6,5))
left, bottom, width, height = 0.1, 0.1, 0.8, 0.8
ax = fig.add_axes([left, bottom, width, height]) 

start, stop, n_values = -8, 8, 800

x_vals = np.linspace(start, stop, n_values)
y_vals = np.linspace(start, stop, n_values)
X, Y = np.meshgrid(x_vals, y_vals)


Z = np.sqrt(X**2 + Y**2)

cp = pl.contourf(X, Y, Z)
pl.colorbar(cp)

ax.set_title('Contour Plot')
ax.set_xlabel('x (cm)')
ax.set_ylabel('y (cm)')
pl.show()
```

```python
import matplotlib.pyplot as pl
pl.style.use('classic')
import numpy as np
```

```python
x=np.random.randn(1000)
pl.hist(x)
```

```python
ax=pl.axes()
ax.set_axisbelow(True)
pl.grid(color='w',linestyle='solid')
for spine in ax.spines.values():
    spine.set_visible(False)
ax.xaxis.tick_bottom()
ax.yaxis.tick_left()
ax.tick_params(colors='gray',direction='out')
for tick in ax.get_xticklabels():
    tick.set_color('gray')
for tick in ax.get_yticklabels():
    tick.set_color('gray')
ax.hist(x,edgecolor='#E6E6E6', color='#EE6666')
pl.show()
```

```python
from matplotlib import cycler
colors=cycler('color',['#EE6666','#3388BB', '#9988DD',
'#EECC55', '#88BB44', '#FFBBBB'])
pl.rc('axes',facecolor='#E6E6E6',edgecolor='none',axisbelow=True,grid=True,prop_cycle=colors)
pl.rc('grid', color='w', linestyle='solid')
pl.rc('xtick', direction='out', color='gray')
pl.rc('ytick', direction='out', color='gray')
pl.rc('patch', edgecolor='#E6E6E6')
pl.rc('lines', linewidth=2)
pl.hist(x)
```

```python
index = pd.MultiIndex.from_product([[2013, 2014], [1, 2]],
names=['year', 'visit'])
columns = pd.MultiIndex.from_product([['Bob', 'Guido', 'Sue'], ['HR', 'Temp'],['Risu','Nisu']],
names=['subject', 'type','Pyara'])
```

```python
import numpy as np
data=np.random.randint(1,50,(4,12))
data
```

```python
data[:,::2]*=10
data+=37
```

```python
health_data=pd.DataFrame(data,index=index,columns=columns)
health_data
```

```python
data_mean=health_data.mean(level='year')
data_mean
```

```python
data_mean.mean(axis=1,level='Pyara')
```

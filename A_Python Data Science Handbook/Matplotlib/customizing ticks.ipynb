```python
ax=pl.axes()
ax.plot(np.random.rand(50))
ax.yaxis.set_major_locator(pl.NullLocator())
ax.xaxis.set_major_formatter(pl.NullFormatter())
```

```python
fig,ax=pl.subplots(5,5,figsize=(5,5))
fig.subplots_adjust(hspace=0,wspace=0)
from sklearn.datasets import fetch_olivetti_faces
faces=fetch_olivetti_faces().images
for i in range(5):
    for j in range(5):
        ax[i,j].xaxis.set_major_locator(pl.NullLocator())
        ax[i,j].yaxis.set_major_locator(pl.NullLocator())
        ax[i,j].imshow(faces[10*i+j],cmap='bone')
```

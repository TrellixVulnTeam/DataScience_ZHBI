```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pl
import seaborn as sns
import plotly.graph_objs as go
from subprocess import check_output
import matplotlib.dates as mdates
```

```python
df_bitc=pd.read_csv('coin_Bitcoin.csv')
df_eth=pd.read_csv('coin_Ethereum.csv')

df_bitc.head()
```

```python
df=df_bitc.copy()
fig=go.Figure()
fig.add_trace(go.Scatter(y=df['High'],
                        x=df['Date'],
                        name='High'))

fig.add_trace(go.Scatter(y=df['Low'],
                        x=df['Date'],
                        name='Low'))
fig.update_layout(title='High and Low stock prices',
                 title_x=.5)
fig.show()
```

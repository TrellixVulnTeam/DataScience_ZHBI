```python
import numpy as np
import matplotlib.pyplot as pl
import pandas as pd
from numpy import nan as na
```

```python
data=np.arange(10)
data
pl.plot(data)
```

```python
fig=pl.figure()
ax1=fig.add_subplot(2,2,1)
ax2=fig.add_subplot(2,2,2)
ax3=fig.add_subplot(2,2,3)
ax4=fig.add_subplot(2,2,4)
ax1.plot(data)
pl.show()
```

```python
fig,ax=pl.subplots(3)
ax[0].plot(data)
ax[1].plot(data)
ax[2].plot(data)
```

```python
fig = pl.figure(figsize=(10,10))
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax1.hist(np.random.randn(100),bins=20,color='k',alpha=0.3)
ax2.scatter(np.arange(30),np.arange(30)+3*np.random.randn(30))
pl.plot(np.random.randn(50).cumsum(),'k--')
```

```python
fig,ax=pl.subplots(nrows=2,ncols=3)
ax
# or directly(subplots(2,3))
```

```python
fig,axes=pl.subplots(2,2,sharex=True,sharey=True)
for i in range(2):
    for j in range(2):
        axes[i,j].hist(np.random.randn(500),bins=20,color='k',alpha=0.5)
pl.subplots_adjust(wspace=0,hspace=0)
```

```python
from numpy.random import randn
pl.figure(figsize=(8,8))
pl.plot(randn(30).cumsum(),'ko:')
```

```python
pl.figure(figsize=(8,8))
pl.plot(randn(30).cumsum(),color='k',linestyle='dotted',marker='o')
```

```python
data=randn(30).cumsum()
pl.figure(figsize=(8,8))
pl.plot(data,'k--',label='Default')
pl.plot(data,'k-',drawstyle='steps-post',label='steps-post')
pl.legend(loc='best')
```

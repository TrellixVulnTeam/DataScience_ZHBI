```python
import numpy as np
import matplotlib.pyplot as pl
```

```python
mean=[0,0]
cov=[[1,1],[1,2]]
x,y=np.random.multivariate_normal(mean,cov,10000).T
```

```python
# squared shape binnings
pl.hist2d(x,y,bins=30,cmap='Blues')
pl.colorbar().set_label('counts in bin')
pl.xlabel("X axis")
pl.ylabel("Y axis")
pl.show()
```

```python
# These are hexagonal binnings
pl.hexbin(x,y,gridsize=30,cmap='Blues')
pl.colorbar().set_label('Counts in bin')
pl.show()
```

# Kernel density Estimation

```python
from scipy.stats import gaussian_kde
data=np.vstack([x,y])
kde=gaussian_kde(data)
```

```python
xgrid=np.linspace(-3.5,3.5,40)
ygrid=np.linspace(-6,6,40)
Xgrid,Ygrid=np.meshgrid(xgrid,ygrid)
Z=kde.evaluate(np.vstack([Xgrid.ravel(),Ygrid.ravel()]))
```

```python
pl.imshow(Z.reshape(Xgrid.shape),origin='lower',aspect='auto',
         extent=[-3.5,3.5,-6,6],cmap='Blues')
pl.colorbar().set_label('Density')
```

```python
pl.style.use('classic')
x=np.linspace(0,10,1000)
fig,ax=pl.subplots()
ax.plot(x,np.sin(x),'-b',label='Sine')
ax.plot(x,np.cos(x),'--r',label='Cosine')
pl.legend(loc='lower left',shadow=True,borderpad=1)
ax.axis('equal')
pl.show()
# borderpad is size of legend box
# we can use upper left lower left upper right lower right and best
```

```python
y=np.sin(x[:,np.newaxis]+np.pi*np.arange(0,2,0.5))
lines=pl.plot(x,y)
pl.legend(lines[:2],['first','second'])
# here lines specifies number of lines in legend box
```

```python
import pandas as pd
```

```python
df = pd.read_json('floorsheet_calculate.txt')
```

```python
df.head()
```

```python
df1 = df.groupby('buyer_broker')[['amount']].sum().sort_values("amount",ascending=False).reset_index().rename(columns = {'buyer_broker':'broker'})
df2 = df.groupby('seller_broker')[['amount']].sum().sort_values("amount",ascending=False).reset_index().rename(columns = {'seller_broker':'broker'})
```

```python
df1.head()
```

```python
df2.head()
```

```python
pd.merge(df1,df2,on = "broker")
```

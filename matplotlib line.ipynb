```python
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(0,10,1)
y=2*x+5

x1=np.arange(0,5,0.5)
y1=2*x+10

fig=plt.figure(figsize=(12,5))
plt.plot(x,y,linewidth=2,linestyle="--",color="m",marker='o',alpha=1)

#we can use linestyle=":" also to print the dotted lines

#Scatter function is used to scatter the pointers 
plt.scatter(x1,y1)

plt.title("Line plot demo")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend(['line1','line2',],loc='best')

# grid is used to make boxes inside the plotted graph
# plt.grid(True)

plt.show()
```

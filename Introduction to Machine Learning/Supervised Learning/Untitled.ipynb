```python
from sklearn.model_selection import train_test_split
X,y=mglearn.datasets.make_forge()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0,test_size=0.3)
```

```python
X_train.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf=KNeighborsClassifier(n_neighbors=3)
```

```python
clf.fit(X_train,y_train)
```

```python
clf.predict(X_test)
```

```python
clf.score(X_test,y_test)
```

```python
fig,axes=pl.subplots(1,3,figsize=(10,3))
for n_neghbots,ax in zip([1,3,9],axes):
    clf=KNeighborsClassifier(n_neighbors=n_neghbots).fit(X,y)
    mglearn.plots.plot_2d_separator(clf,X,fill=True,eps=0.1,ax=ax,alpha=0.3)
    mglearn.discrete_scatter(X[:,0],X[:,1],y,ax=ax)
    ax.set_title("{}".format(n_neghbots))
    ax.set_xlabel("Feature 0")
    ax.set_ylabel("Feature 1")
```

```python
fig,axes=pl.subplots(1,3,figsize=(10,3))
for n,ax in zip([1,3,9],axes):
    clf=KNeighborsClassifier(n_neighbors=n).fit(X,y)
    mglearn.plots.plot_2d_separator(clf,X,fill=True,ax=ax,alpha=0.3,eps=0.2)
    mglearn.discrete_scatter(X[:,0],X[:,1],y,ax=ax)
    ax.set_title(n)
    ax.set_xlabel("Feature 0")
    ax.set_ylabel("Feature 1")
axes[0].legend()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
cancer.data, cancer.target, stratify=cancer.target, random_state=66)
```

```python
training_accuracy=[]
test_accuracy=[]
neighbors_settings=range(1,11)
for n in neighbors_settings:
    clf=KNeighborsClassifier(n_neighbors=n).fit(X_train,y_train)
    training_accuracy.append(clf.score(X_train,y_train))
    test_accuracy.append(clf.score(X_test,y_test))
pl.plot(neighbors_settings,training_accuracy,label="Training accuracy")
pl.plot(neighbors_settings,test_accuracy,linestyle='dashed',label="Test Accuracy")
pl.legend()
```

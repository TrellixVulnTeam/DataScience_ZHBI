```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as pl
import numpy as np
digits=load_digits()
```

```python
a=np.random.randint(1,20,20)
b=np.random.randint(1,20,20)
m_a=np.mean(a)
m_b=np.mean(b)
```

```python
pl.scatter(a,b)
```

```python
a=a-m_a
b=b-m_b
```

```python
pl.scatter(a,b)
```

```python
fig,axes=pl.subplots(2,5,figsize=(10,5),subplot_kw={'xticks':(),
                                                    'yticks':()})
for ax,img in zip(axes.ravel(),digits.images):
    ax.imshow(img)
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
pca.fit(digits.data)
digits_pca=pca.transform(digits.data)
colors = ["#476A2A", "#7851B8", "#BD3430", "#4A2D4E", "#875525",
"#A83683", "#4E655E", "#853541", "#3A3120", "#535D8E"]
pl.figure(figsize=(10,10))
pl.xlim(digits_pca[:,0].min(),digits_pca[:,0].max())
pl.ylim(digits_pca[:,1].min(),digits_pca[:,1].max())
for i in range(len(digits.data)):
    pl.text(digits_pca[i,0],digits_pca[i,1],str(digits.target[i]),
           color=colors[digits.target[i]],
           fontdict={'weight':'bold','size':9})
```

```python
from sklearn.manifold import TSNE
tsne=TSNE(random_state=42)
digits_tsne=tsne.fit_transform(digits.data)
```

```python
pl.figure(figsize=(10,10))
pl.xlim(digits_tsne[:, 0].min(), digits_tsne[:, 0].max() + 1)
pl.ylim(digits_tsne[:, 1].min(), digits_tsne[:, 1].max() + 1)
for i in range(len(digits.data)):
    pl.text()


```

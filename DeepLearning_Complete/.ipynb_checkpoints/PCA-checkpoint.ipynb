```python
A=np.array([[1,2],[3,4],[5,6]])
A
```

```python
M=np.mean(A.T,axis=1)
M
```

```python
C=A-M
C
```

```python
V=np.cov(C.T)
V
```

```python
val,vec=np.linalg.eig(V)
```

```python
val,vec
```

```python
P=np.dot(C,vec)
```

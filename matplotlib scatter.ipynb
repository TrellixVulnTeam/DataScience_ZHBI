```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
a=[1,5,2,6,3,7,1,8,3,9] 
b=[6,1,3,8,4,9,4,6,1,8]
x=[4,5,3,8,9,5,6,1,7,2]
plt.scatter(a,x,color="g",edgecolors="black",marker='o',s=200)
plt.scatter(x,b,color='#0000FF',marker='4',s=200)
# we can have marker value as 1 2 3 4 also so the arrows will be shown in different directions

plt.grid()
plt.title("Demo")
plt.xlabel('X axis')
plt.legend(['first','second'],loc='best')
plt.ylabel('Y axis')
plt.show()
```

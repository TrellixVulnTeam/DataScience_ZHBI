```python
import pandas as pd
import matplotlib.pyplot as pl
data=pd.read_csv("brain.csv")
dataa=data.loc[:,["size","wt"]]
pl.figure(figsize=(15,8))
pl.scatter(dataa["size"],dataa["wt"])
pl.legend(["Points"],loc="best")
pl.title("Head and Brain size prediction")
pl.xlabel("Head size")
pl.ylabel("Brain weight ")
pl.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

x=pd.DataFrame(dataa["size"])
y=pd.DataFrame(dataa["wt"])

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)

regresser=LinearRegression()

regresser.fit(x_train,y_train)

# regresser.fit(x,y)

print(regresser.intercept_)
print(regresser.coef_)
print()
y_pred=regresser.predict(x_test)
y_prd=pd.DataFrame(y_pred,columns=['Predicted'])
print(regresser.score(x_train,y_train))
pl.figure(figsize=(15,8))
pl.scatter(y_test,y_prd)
```

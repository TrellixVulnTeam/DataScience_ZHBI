```python
import mglearn
import numpy as np
import pandas as pd
import matplotlib.pyplot as pl
import seaborn as sns
```

```python
X,y=mglearn.datasets.make_forge()
mglearn.discrete_scatter(X[:,0],X[:,1],y)
pl.legend(['Class 0','Class 1'],loc=4)
pl.xlabel("First Feature")
pl.ylabel("Second Feature")
X.shape
```

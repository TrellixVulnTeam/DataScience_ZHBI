```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as pl
```

```python
def f(x,y):
    return np.sin(np.sqrt(x**2+y**2))
x=np.linspace(-6,6,30)
y=np.linspace(-6,6,30)
X,Y=np.meshgrid(x,y)
Z=f(X,Y)
```

```python
fig=pl.figure()
ax=pl.axes(projection='3d')
ax.contour3D(X,Y,Z,50,cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(40,10)
```

```python
fig=pl.figure()
ax=pl.axes(projection='3d')
ax.plot_wireframe(X,Y,Z,color='black')
```

```python
ax=pl.axes(projection='3d')
ax.plot_surface(X,Y,Z,rstride=2,cstride=2,cmap='viridis',edgecolor='none')
```

```python
r=np.linspace(0,6,20)
theta=np.linspace(-0.9*np.pi,0.8*np.pi,40)
r,theta=np.meshgrid(r,theta)
X=r*np.sin(theta)
Y=r*np.cos(theta)
Z=f(X,Y)
ax=pl.axes(projection='3d')
ax.plot_surface(X,Y,Z,cmap='viridis')
```

```python
theta=2*np.pi*np.random.random(1000)
r=6*np.random.random(1000)
x=np.ravel((r*np.sin(theta)))
y=np.ravel((r*np.cos(theta)))
z=f(x,y)
ax=pl.axes(projection='3d')
ax.scatter(x,y,z,c=z,cmap='viridis',edgecolor='black')
```

```python
ax=pl.axes(projection='3d')
ax.plot_trisurf(x,y,z,cmap='viridis',edgecolor='none')
```

```python
theta=np.linspace(0,2*np.pi,30)
w=np.linspace(-0.25,0.25,8)
w,theta=np.meshgrid(w,theta)
phi=0.5*theta
```

```python
r=1+w*np.cos(phi)
x=np.ravel(r*np.cos(theta))
y=np.ravel(r*np.sin(theta))
z=np.ravel(r*np.sin(phi))
```

```python
from matplotlib.tri import Triangulation
tri=Triangulation(np.ravel(w),np.ravel(theta))
ax=pl.axes(projection='3d')
ax.plot_trisurf(x,y,z,triangles=tri.triangles,cmap='viridis',linewidth=0.2)
ax.set_xlim(-1,1)
ax.set_ylim(-1,1)
ax.set_zlim(-1,1)
```

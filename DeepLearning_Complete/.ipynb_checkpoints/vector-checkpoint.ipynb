```python
import numpy as np
import matplotlib.pyplot as pl
from sklearn.datasets import make_circles
X,y=make_circles(n_samples=100)
```

```python
a=np.random.rand(10)
b=np.random.rand(10)
c=[30,60,90,120,150,180,210,240,5,20]
```

```python
pl.scatter(a,b,c=c)
```

```python
m=np.array([[1,2],[1,1]])
```

```python
new_x=np.zeros(10)
new_y=np.zeros(10)
for i in range(len(a)):
    vector=np.array([a[i],b[i]])
    dots=np.dot(vector,m)
    new_x[i]=dots[0]
    new_y[i]=dots[1]
```

```python
pl.scatter(new_x,new_y)
```

```python
X,y=make_circles(n_samples=100)
```

```python
m=np.array([[4,3],[3,6]])
eigval,eigvec=np.linalg.eig(m)
pl.figure(figsize=(5,5))
pl.quiver(0,0,eigvec[0][0],eigvec[1][0],angles='xy',scale_units='xy',
         scale=1)
pl.quiver(0,0,eigvec[0][1],eigvec[1][1],angles='xy',scale_units='xy',
         scale=1)
pl.scatter(X[:,0],X[:,1])
```

```python
new_x=np.dot(X,m)
pl.scatter(new_x[:,0],new_x[:,1])
```

```python
eigval
```

```python
eigvec
```

```python
soa1=np.array([[0,0,eigval[0]*eigvec[0,0],eigval[0]*eigvec[1,0]]])
soa2=np.array([[0,0,eigval[1]*eigvec[0,1],eigval[1]*eigvec[1,1]]])
a,b,c,d=zip(*soa1)
e,f,g,h=zip(*soa2)
pl.scatter(new_x[:,0],new_x[:,1])
pl.quiver(a,b,c,d,angles='xy',scale_units='xy',scale=1)
pl.quiver(e,f,g,h,angles='xy',scale_units='xy',scale=1)
pl.draw()
pl.show()
```

```python
x=np.linspace(-1,1,100)
y1=np.sqrt(1-np.square(x))
y2=-1*y1
eigval,eigvec=np.linalg.eig(np.array([[9,4],[4,3]]))
eigval,eigvec
```

```python
pl.plot(x,y1,'b')
pl.plot(x,y2,'b')
soa1=np.array([[0,0,eigvec[0,0],eigvec[1,0]]])
soa2=np.array([[0,0,eigvec[0,1],eigvec[1,1]]])
a,b,c,d=zip(*soa1)
e,f,g,h=zip(*soa2)
pl.quiver(a,b,c,d,angles='xy',scale_units='xy',scale=1)
pl.quiver(e,f,g,h,angles='xy',scale_units='xy',scale=1)
# pl.xlim([-1,10])
# pl.ylim([-1,10])
pl.draw()
pl.show()
```

```python
def transformation(x,y):
    return 9*x+4*y,4*x+3*y
```

```python
x_new1,y_new1=transformation(x,y1)
x_new2,y_new2=transformation(x,y2)
```

```python
soa1=np.array([[0,0,eigval[0]*eigvec[0,0],eigval[0]*eigvec[1,0]]])
soa2=np.array([[0,0,eigval[1]*eigvec[0,1],eigval[1]*eigvec[1,1]]])
a,b,c,d=zip(*soa1)
e,f,g,h=zip(*soa2)
pl.plot(x_new1,y_new1,'r')
pl.plot(x_new2,y_new2,'r')
pl.quiver(a,b,c,d,angles='xy',scale_units='xy',scale=1)
pl.quiver(e,f,g,h,angles='xy',scale_units='xy',scale=1)
pl.draw()
pl.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pl
```

```python
cities=pd.read_csv('california_cities.csv')
cities=cities.drop(columns={"Unnamed: 0"})
```

```python
cities
```

```python
lat,lon=cities['latd'],cities.longd
pop,area=cities.population_total,cities.area_total_km2
```

```python
pl.scatter(lon,lat,label=None,c=np.log10(pop),cmap="viridis",
          s=area,linewidths=0,alpha=0.5)
pl.axis(aspect='equal')
pl.clim(3,7)
for area in [100,300,500]:
    pl.scatter([],[],c='k',alpha=0.3,s=area,label=str(area)+"km$^2$")
pl.legend(scatterpoints=1,frameon=False, labelspacing=1,
         title='City Area')
```

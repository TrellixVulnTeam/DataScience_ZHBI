```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

#datas:
plt.figure(figsize=(10,5))
total=np.array([20,10,50,40,30,20,10,40,60])
order=np.array([10,20,50,40,70,80,60,30,20])
discount=np.array([90,20,10,80,80,40,12,19])
total.sort()
order.sort()
discount.sort()
print(total,order,discount)
print(np.mean(total),order.mean(),discount.mean())
print(np.median(total),np.median(order),np.median(discount))
data=list([total,order,discount])
plt.boxplot(data,showmeans=True,vert=True)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,5))
total=np.array([20,10,50,40,30,20,10,40,60])
order=np.array([10,20,50,40,70,80,60,30,20])
discount=np.array([90,20,10,80,80,40,12,19])
total.sort()
order.sort()
discount.sort()
print(total,order,discount)
print(np.mean(total),order.mean(),discount.mean())
print(np.median(total),np.median(order),np.median(discount))
data=list([total,order,discount])
plt.violinplot(data,showmeans=True,showmedians=True,vert=True)
plt.grid()
plt.show()
```

```python
labels=["cow","bufallo","chicken","fish"]
value=[20,40,15,20]
fig=plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title("Pie chart 1")
plt.pie(value,labels=labels,shadow=True,autopct='%1.1f%%',startangle=90)

plt.subplot(1,2,2)
plt.title("Pie chart 2")
plt.pie(value,labels=labels,shadow=True,autopct='%1.1f%%',startangle=90,explode=(0.1,0.1,0,0))

plt.show()

```

```python
import numpy as np
import pandas as pd
dfs=pd.read_csv('cars.csv')
dfs
dfs.vs=dfs.vs.fillna(dfs.vs.mean())
dfs
npp=np.array(dfs)
print(npp)
print(np.array_equal(dff,npp))
first=npp[:,8:9]
second=dff[:,8:9]
np.array_equal(first,second)
```

```python
f=lambda x: x-1
dfs['am']=dfs['am'].apply(f)

dfs[dfs['am']==0]
dfs
```

```python
# updating new row 2*am with the value of am by adding 1
apply=lambda x: x+1
dfs['2*am']=dfs['am'].apply(apply)


dfs
```

```python
dfs

#renaming the 1st column
cars=dfs.rename(columns={'Unnamed: 0':'Models'})

# sorting the values by column vs
cars.sort_values(by='vs')
```

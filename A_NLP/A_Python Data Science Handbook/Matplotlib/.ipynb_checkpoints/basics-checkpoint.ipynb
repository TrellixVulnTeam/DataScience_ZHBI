```python
import matplotlib as mpl
import matplotlib.pyplot as pl
pl.style.use('classic')
import numpy as np
```

```python
x=np.linspace(0,10,100)
```

```python
pl.plot(x,np.sin(x))
```

```python
pl.plot(x,np.cos(x))
```

```python
fig=pl.figure()
pl.plot(x,np.sin(x),'-')
pl.plot(x,np.cos(x),'--')
pl.show()
```

```python
fig.savefig('wave.png')
```

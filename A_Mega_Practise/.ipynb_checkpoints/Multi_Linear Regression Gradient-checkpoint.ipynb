```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as pl
import numpy as np
import pandas as pd

X,y=make_regression(n_samples=300,n_features=5,noise=10)
y=y.reshape(-1,1)
```

```python
pl.figure(figsize=(10,6))

pl.subplot(1,2,1)
pl.scatter(X[:,0],y)
pl.subplot(1,2,2)
pl.scatter(X[:,1],y)
```

```python
pl.figure(figsize=(10,10))
pl.subplot(2,2,1)
pl.scatter(X[:,2],y)
pl.subplot(2,2,2)
pl.scatter(X[:,3],y)
pl.subplot(2,2,3)
pl.scatter(X[:,4],y)
```

```python
class classify:
    def __init__(self,L):
        self.L=L
    def predictor(self,epoch):
        epoch=1500
        L=0.01
        M = np.random.rand(5,1)
        c=0
        n=X.shape[0]
        err=[]
        for i in range(epoch):
            ypred=np.matmul(X,M)+c
            d_m=(-2/n)*np.sum(X*(y-ypred),axis=0)
            d_c=(-2/n)*np.sum(y-ypred)
            M=M-L*d_m
            c=c-L*d_c
            error=np.mean(np.square(np.subtract(y,ypred)))
            err.append(error)
        print(np.mean(M,axis=1),c)
        pl.plot(err)
c=classify(0.01)
c.predictor(100)
```

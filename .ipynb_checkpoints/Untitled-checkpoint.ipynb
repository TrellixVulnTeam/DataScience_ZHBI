```python
import numpy as np
import matplotlib.pyplot as pl
```

```python
a=np.arange(100)
b=np.random.randint(1,100,100)
```

```python
pl.scatter(a,b)
pl.plot(a,2*a+2)
pl.plot(a,0.2*a+0.2)
pl.plot(a,4*a+4)
```

```python
import pandas as pd
pop = pd.read_csv('state-population.csv')
areas = pd.read_csv('state-areas.csv')
abbrevs = pd.read_csv('state-abbrevs.csv')
```

```python
print(pop.shape)
pop.head()
```

```python
print(areas.shape)
areas.head()
```

```python
print(abbrevs.shape)
abbrevs.head()
```

```python
merged=pd.merge(pop,abbrevs,how="outer",left_on='state/region',
               right_on='abbreviation')
merged
```

```python
merged=merged.drop(columns={'abbreviation'})
# or merged=merged.drop("abbreviation",1)
merged
```

```python
merged.isnull().any()
```

```python
merged[merged["population"].isnull()].head()
```

```python
a=merged.loc[merged["state"].isnull(),'state/region'].unique()
a
```

```python
merged.loc[merged['state/region']=='PR','state']="Puerto Rico"
merged.loc[merged['state/region']=='USA','state']="United States"

# Or we can implement this one too
# merged.state[merged["state/region"]=='USA']="United States"
# merged.state[merged["state/region"]=='PR']="Puerto Rico"

merged
```

```python
merged.info()
```

```python
final=pd.merge(merged,areas,on="state",how="left")
final
```

```python
print(final.info())
final.isnull().any()
```

```python
final.shape
```

```python
final.dropna(inplace=True)
# inplace function is used to save the state of dropped null values
# if we dont give inplace, than the rows will be dropped only in view
# but will not be saved
```

```python
final.head()
```

```python
print(2544-2476,"Rows are deleted with null values")
final.shape
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pl
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
```

```python
x,y=make_regression(n_samples=1000,n_features=3,noise=10)
df=pd.DataFrame(x)
df["Target"]=pd.DataFrame(y)
```

```python
dff=df.rename(columns={0:"A",
                      1:"B",
                      2:"C"})
dff
```

```python
pl.figure(figsize=(15,8))
for i in range(3):
    pl.scatter(dff.iloc[:,i],dff["Target"])
print(x.shape)
y=y.reshape(-1,1)
print(y.shape)
```

```python
m=np.zeros(3).reshape(-1,1)
c=0
L=0.01
reps=1000
n=1000
error=[]
d_m=[0,0,0]
d_c=0
for i in range(reps):
    y_pred=np.dot(x,m)+c
    d_m=(-2/n)*np.sum((y-y_pred)*x,axis=0).reshape(-1,1)
    d_c=(-2/n)*sum(y-y_pred)
    m=m-L*d_m
    c=c-L*d_c
    er=np.mean(np.square(y-y_pred))
    error.append(er)
print(m,c)
```

```python
pl.plot(error)
```

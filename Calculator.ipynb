```python
import tkinter as tk

root= tk.Tk()

canvas1 = tk.Canvas(root, width = 400, height = 500)
canvas1.pack()

entry1 = tk.Entry (root) 
canvas1.create_window(260, 100, window=entry1,width=250,height=30)

entry2 = tk.Entry (root) 
canvas1.create_window(260, 140, window=entry2,width=250,height=30)

entry3 = tk.Entry (root) 
canvas1.create_window(260, 240, window=entry3,width=250,height=30)

label0 = tk.Label(root, text='Calculator')
label0.config(font=('TimesnewRoman', 14))
canvas1.create_window(150, 40, window=label0)

label1 = tk.Label(root, text=' Value 1:')
label1.config(font=('TimesnewRoman', 10))
canvas1.create_window(100, 100, window=label1)

label2 = tk.Label(root, text=' Value 2:')
label2.config(font=('TimesnewRoman', 10))
canvas1.create_window(100, 140, window=label2)

label3 = tk.Label(root, text='Result:')
label3.config(font=('TimesnewRoman', 10))
canvas1.create_window(100, 240, window=label3)
label_temp=tk.Label(root,text="")
canvas1.create_window(310, 240, window=label_temp)

def add():  
    try:
        v1 = entry1.get()
        v2 = entry2.get()
        
        label4 = tk.Label(root, text= float(v1)+float(v2),font=('helvetica', 10, 'bold'),bg='white')
        canvas1.create_window(210, 240, window=label4)
        label_temp=tk.Label(root,text="                                                            ")
        canvas1.create_window(210, 300, window=label_temp)
    except:
        label_temp=tk.Label(root,text="Please take numeric values")
        canvas1.create_window(210, 300, window=label_temp)
        label4 = tk.Label(root, text="          ",font=('helvetica', 10, 'bold'),bg='white')
        canvas1.create_window(210, 240, window=label4)

        
      
buttonAdd = tk.Button(text='+', command=add, bg='green', fg='white', font=('helvetica', 9, 'bold'), width = 5)
canvas1.create_window(90, 190, window=buttonAdd)

def sub():  
    try:
        v1 = entry1.get()
        v2 = entry2.get()
        
        label4 = tk.Label(root, text= float(v1)-float(v2),font=('helvetica', 10, 'bold'),bg='white')
        canvas1.create_window(210, 240, window=label4)
        label_temp=tk.Label(root,text="                                                            ")
        canvas1.create_window(210, 300, window=label_temp)
    except:
        label_temp=tk.Label(root,text="Please take numeric values")
        canvas1.create_window(210, 300, window=label_temp)
        label4 = tk.Label(root, text="          ",font=('helvetica', 10, 'bold'),bg='white')
        canvas1.create_window(210, 240, window=label4)
      
buttonSub = tk.Button(text='â€“', command=sub, bg='green', fg='white', font=('helvetica', 9, 'bold'), width = 5)
canvas1.create_window(140, 190, window=buttonSub)

def mul():  
    try:
        v1 = entry1.get()
        v2 = entry2.get()
        
        label4 = tk.Label(root, text= float(v1)*float(v2),font=('helvetica', 10, 'bold'),bg='white')
        canvas1.create_window(210, 240, window=label4)
        label_temp=tk.Label(root,text="                                                            ")
        canvas1.create_window(210, 300, window=label_temp)
    except:
        label_temp=tk.Label(root,text="Please take numeric values")
        canvas1.create_window(210, 300, window=label_temp)
        label4 = tk.Label(root, text="          ",font=('helvetica', 10, 'bold'),bg='white')
        canvas1.create_window(210, 240, window=label4)
      
buttonMul = tk.Button(text='x', command=mul, bg='green', fg='white', font=('helvetica', 9, 'bold'), width = 5)
canvas1.create_window(190, 190, window=buttonMul)

def div():  
    try:
        v1 = entry1.get()
        v2 = entry2.get()
        
        label4 = tk.Label(root, text= float(v1)/float(v2),font=('helvetica', 10, 'bold'),bg='white')
        canvas1.create_window(210, 240, window=label4)
        label_temp=tk.Label(root,text="                                                            ")
        canvas1.create_window(210, 300, window=label_temp)
    except:
        label_temp=tk.Label(root,text="Please take numeric values")
        canvas1.create_window(210, 300, window=label_temp)
        label4 = tk.Label(root, text="          ",font=('helvetica', 10, 'bold'),bg='white')
        canvas1.create_window(210, 240, window=label4)
      
buttonDiv = tk.Button(text='/', command=div, bg='green', fg='white', font=('helvetica', 9, 'bold'), width = 5)
canvas1.create_window(240, 190, window=buttonDiv)

def clear():  
    v1 = entry1.get()
    v2 = entry2.get()
    entry1.delete(0,"end")
    entry2.delete(0,"end")
    label8 = tk.Label(root, text= "",font=('helvetica', 10, 'bold'),bg='white')
    canvas1.create_window(210, 240, window=label8)
    label4 = tk.Label(root, text="          ",font=('helvetica', 10, 'bold'),bg='white')
    canvas1.create_window(210, 240, window=label4)
    
clear = tk.Button(text='Clear', command=clear, bg='green', fg='white', font=('helvetica', 9, 'bold'), width = 5)
canvas1.create_window(290, 190, window=clear)


root.mainloop()
```

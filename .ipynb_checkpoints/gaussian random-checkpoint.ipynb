```python
# import numpy as np
import matplotlib.pyplot as pl

def models(a):
    formula1=((1/(2*np.pi*a))*np.exp(-(a-np.mean(a))**2)/2*(a**2))
    formula2=np.sin(a)
    plot(formula1,formula2)
def plot(formula1,formula2):
    pl.figure(figsize=(10,8))
    pl.plot(formula1,linestyle=':',linewidth=3)
    pl.plot(formula2,linestyle='--',linewidth=3)
    pl.legend(["Gaussian","Sin wave"])
    
    pl.show()
a=np.arange(1,100)
models(a)
```

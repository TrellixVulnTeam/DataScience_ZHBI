```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as pl
```

```python
data=sns.load_dataset('iris')
```

```python
data.head()
```

```python
sns.pairplot(data,hue='species')
```

```python
sns.boxplot(data=data,y='petal_length',x='species')
```

```python
sns.boxplot(data=data,y='sepal_length',x='species')
```

```python
sns.boxplot(data=data,y='sepal_width',x='species')
```

```python
df_setosa=data.loc[data['species']=='setosa']
df_virginica=data.loc[data['species']=='virginica']
df_versicolor=data.loc[data['species']=='versicolor']
import numpy as np
```

```python
pl.plot(df_setosa['sepal_length'],np.zeros_like(df_setosa['sepal_length']),'o')
pl.plot(df_virginica['sepal_length'],np.zeros_like(df_virginica['sepal_length']),'o')
pl.plot(df_versicolor['sepal_length'],np.zeros_like(df_versicolor['sepal_length']),'o')
pl.xlabel('Sepal Length')
pl.show()
```

```python
pl.figure(figsize=(10,8))
sns.FacetGrid(data,hue='species',size=6)\
    .map(pl.scatter,'petal_length','petal_width')\
    .add_legend()
```

```python
import numpy as np
import matplotlib.pyplot as pl
from sklearn.datasets import make_blobs
X,y=make_blobs(n_samples=1000,n_features=2,centers=4,random_state=0)
from sklearn.model_selection import train_test_split
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
y=y%2
```

```python
pl.figure(figsize=(10,7))
pl.scatter(X[:,0],X[:,1],c=y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,stratify=y, random_state=0)
from sklearn.metrics import mean_squared_error
```

```python
model=Sequential()
model.add(Dense(input_dim=2,units=20,activation='sigmoid'))
model.add(Dense(units=1,activation='sigmoid'))
model.compile(optimizer='adam',loss='mean',metrics=['mse'])
```

```python
model.summary()
```

```python
h=model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10000)
```

```python
pl.plot(h.history['loss'])
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as pl
pd.set_option('display.max_columns',None)
```

```python
train=pd.read_csv('train.csv')
```

In Data analysis we will analyze to find out below stuffs:
1. Missing values
2. Numerical values
3. Distribution of numerical values

4. Categorical values
5. Cardinality of Categorical Variables
6. Ouliers

7. Relationship between independent and dependent feature(SalePrice)

```python
features_withna=[columns for columns in train.columns if train[columns].isnull().sum()>0]
for columns in features_withna:
    a=train[columns].isnull().sum()
    total=np.round(a/len(train[columns]),4)
    print("Missing values in ",columns,"are",total,"%")
```

```python
sns.heatmap(train[features_withna].isnull())
```

```python
for feature in features_withna:
    data=train.copy()
    data[feature]=np.where(data[feature].isnull(),1,0)
    data.groupby(feature)['SalePrice'].median().plot.bar()
    pl.title(feature)
    pl.show()
```

For most of the features, Sale Price is playing independent roles with
respect to null values

```python
data=train.copy()
```

```python
numeric_features=[columns for columns in data.columns if data[columns].dtype!='O']
print("number of numeric values are",len(numeric_features))
data[numeric_features].head()
```

```python
pd.Series(['A']).dtype,pd.Series([1]).dtype,pd.Series([0.5]).dtype,
```

```python
features_with_year=[columns for columns in numeric_features if 'Yr' in columns or 'Year' in columns or 'yr' in columns or 'year' in columns]
features_with_year
```

```python
for feature in features_with_year:
    print(feature,np.unique(data[feature]))
```

```python
for features in features_with_year:
    data.groupby(features)['SalePrice'].median().plot(color='c')
    pl.ylabel("Sales Price")
    pl.title(features)
    pl.show()
```

```python
for features in features_with_year:
    if features != 'YrSold':
        temp=train.copy()
        temp[feature]=temp['YrSold']-temp[features]
        pl.scatter(temp[feature],temp['SalePrice'])
        pl.title(features)
        print(np.max(temp[feature]))
        pl.ylabel("Sales Price")
        pl.xlabel("Years old")
        pl.show()
```

```python
train[features_with_year]
```

```python
discrete=[i for i in train[numeric_features].columns if len(np.unique(train[i]))<25 and i not in features_with_year]
```

```python
discrete
```

```python
train[discrete]
```

```python
for feature in discrete:
    data=train.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    pl.title(feature)
    pl.show()
```

```python
continous=[feature for feature in numeric_features if feature not in discrete and feature not in features_with_year+['Id']]
```

```python
len(continous)
```

```python
for feature in continous:
    data=train.copy()
    data[feature].hist(bins=25)
    pl.title(feature)
    pl.ylabel('Sales Price')
    pl.show()
```

Since the data is not normally distributed so we will be using 
logarithmic transformation

```python
data=train.copy()
for feature in continous:
    pl.scatter(data[feature],data['SalePrice'])
    pl.title(feature)
    pl.show()
```

```python
data=train.copy()
for feature in continous:
    data[feature]=np.log(data[feature])
    pl.scatter(data[feature],data['SalePrice'])
    pl.title(feature)
    pl.show()
    
```

```python
data=train.copy()
for feature in continous:
    pl.hist(data[feature],bins=25)
    pl.title(feature)
    pl.show()
```

```python
data=train.copy()
for feature in continous:
    data[feature]=np.log(data[feature])
    data[data[feature]==-np.inf]=np.nan
    data[feature].dropna(inplace=True)
    if len(data[feature])>0:
        pl.hist(data[feature],bins=25)
        pl.title(feature)
        pl.show()
```

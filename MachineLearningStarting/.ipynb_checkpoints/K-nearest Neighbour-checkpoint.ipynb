```python
import numpy as np
import pandas as pd
import mglearn
import matplotlib.pyplot as pl
import seaborn as sns
import scipy
```

```python
from sklearn.datasets import make_blobs
X,y=make_blobs(n_samples=50,n_features=2,centers=2,cluster_std=2)
df=pd.DataFrame(X,columns=['X1','X2'])
df['y']=y
```

```python
import seaborn as sns
sns.FacetGrid(df,hue='y')\
    .map(pl.scatter,'X1','X2')\
    .add_legend()
```

```python
def predict(data,neighbors):
    dist=[]
    neighbor_datas=[]
    for i in range(len(X)):
        distance=np.sqrt(np.square(data[0]-X[:,0][i])+np.square(data[1]-X[:,1][i]))
        dist.append(distance)
    distance_prediction=dist.copy()
    dists=[]
    for i in range(neighbors):
        minimum=np.argmin(distance_prediction)
        dists.append(minimum)
        distance_prediction[minimum]=999
    print(y[dists])
    return (scipy.stats.mode(y[dists])[0])
    


neighbors=5
point=np.array([-6,-6])
label=predict(point,neighbors)
print(label)


sns.FacetGrid(df,hue='y',size=7)\
    .map(pl.scatter,'X1','X2')\
    .add_legend()
if label==0:
    pl.scatter(point[0],point[1],marker='*',s=200,c='blue')
else:
    pl.scatter(point[0],point[1],marker='*',s=200,c='orange')
```

```python
import matplotlib.pyplot as pl
# pl.style.use("seaborn-whitegrid")
import numpy as np
```

```python
x=np.linspace(0,10,1000)
pl.plot(x,np.sin(x))
pl.xlim(-1,11)
pl.ylim(-1.5,1.5)
```

```python
# this rotated the axis
pl.plot(x,np.sin(x))
pl.xlim(10,0)
pl.ylim(-1.2,1.2)
```

```python
pl.axis([-1,11,-1.5,1.5])
pl.plot(x,np.sin(x))
```

```python
pl.plot(x,np.sin(x))
pl.axis('tight')
```

```python
pl.plot(x,np.sin(x))
pl.axis('equal')
```

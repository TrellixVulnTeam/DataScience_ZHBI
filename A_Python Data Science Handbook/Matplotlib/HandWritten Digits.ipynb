```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as pl
digits=load_digits()
```

```python
fig,ax=pl.subplots(8,8,figsize=(8,8))
for i,ax in enumerate(ax.flat):
    ax.imshow(digits.images[i],cmap='binary')
    ax.set(xticks=[],yticks=[])
```

```python
for i in range(5):
    pl.imshow(digits.images[i])
    pl.show
```

```python
from sklearn.manifold import Isomap
iso=Isomap(n_components=4)
projection=iso.fit_transform(digits.data)
pl.scatter(projection[:,0],projection[:,3],lw=0.1,c=digits.target,
          cmap=pl.cm.get_cmap('cubehelix',6))
pl.colorbar(ticks=range(6),label='digit value')
```

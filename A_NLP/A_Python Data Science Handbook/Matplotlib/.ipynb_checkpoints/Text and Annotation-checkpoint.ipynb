```python
import matplotlib.pyplot as pl
import matplotlib as mpl
pl.style.use('seaborn-whitegrid')
import numpy as np
import pandas as pd
import seaborn as sns
import datetime
```

```python
births=pd.read_csv('births.csv')
births.head()
```

```python
sns.FacetGrid(births,hue="year")\
    .map(pl.scatter,"month","day")\
    .add_legend()
```

```python
quartiles=np.percentile(births['births'],[25,50,75])
# print("min is",np.min(births.births))
# print("max is",np.max(births.births))
# quartiles
# births.dropna(inplace=True)
# births.shape
```

```python
mu,sig=quartiles[1],0.74*(quartiles[2]-quartiles[0])
births=births.query('(births>@mu-5*@sig)&(births<@mu+5*@sig)')
births['day']=births['day'].astype(int)
```

```python
births.shape
```

```python
births.index=pd.to_datetime(10000*births.year+
                            100*births.month+
                            births.day,format='%Y%m%d'
                           )
```

```python
births.head()
```

```python
births_bydate=births.pivot_table('births',
                                 [births.index.month,births.index.day])
births_bydate.index = [pd.datetime(2012, month, day) for (month, day) in births_bydate.index]
births_bydate.head()
```

```python
fig,ax=pl.subplots(figsize=(12,4))
births_bydate.plot(ax=ax)
```

```python
fig, ax = pl.subplots(figsize=(12, 4))
births_bydate.plot(ax=ax)
# Add labels to the plot
style = dict(size=10, color='gray')
ax.text('2012-1-1', 3950, "New Year's Day", **style)
ax.text('2012-7-4', 4250, "Independence Day", ha='center', **style)
ax.text('2012-9-4', 4850, "Labor Day", ha='center', **style)
ax.text('2012-10-31', 4600, "Halloween", ha='right', **style)
ax.text('2012-11-25', 4450, "Thanksgiving", ha='center', **style)
ax.text('2012-12-25', 3850, "Christmas ", ha='right', **style)
# Label the axes
ax.set(title='USA births by day of year (1969-1988)',
ylabel='average daily births')
# Format the x axis with centered month labels
ax.xaxis.set_major_locator(mpl.dates.MonthLocator())
ax.xaxis.set_minor_locator(mpl.dates.MonthLocator(bymonthday=15))
ax.xaxis.set_major_formatter(pl.NullFormatter())
ax.xaxis.set_minor_formatter(mpl.dates.DateFormatter('%h'));
```

```python
fig,ax=pl.subplots(facecolor='lightgray')
ax.axis([0,10,0,10])
ax.text(1,5,".Data:(1,5)",transform=ax.transData)
ax.text(0.5,0.1,".Axes:(0.5,0.1)",transform=ax.transAxes)
ax.text(0.2,0.2,".Figure(0.2,0.2)",transform=fig.transFigure)
```

```python
import matplotlib.pyplot as pl
import numpy as np
```

```python
x=np.linspace(0,10,100)
pl.subplot(3,1,1)
pl.plot(x,np.sin(x))
pl.subplot(3,1,2)
pl.plot(x,np.cos(x))
pl.subplot(3,1,3)
pl.plot(x,np.tan(x))
```

```python
pl.subplot(3,1,1)
pl.plot(x,np.cosh(x))
pl.subplot(3,1,2)
pl.plot(x,np.sinh(x))
pl.subplot(3,1,3)
pl.plot(x,np.tanh(x))
```

```python
fig,ax=pl.subplots(3)
pl.figure(figsize=(10,8))
ax[0].plot(x,np.sin(x))
ax[1].plot(x,np.cos(x))
ax[2].plot(x,np.tan(x))
```

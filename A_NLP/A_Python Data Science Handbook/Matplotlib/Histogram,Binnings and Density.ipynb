```python
import numpy as np
import matplotlib.pyplot as pl
pl.style.use('seaborn-white')
```

```python
data=np.random.randn(1000)
data=data.reshape(-1,1)
```

```python
pl.hist(data)
```

```python
pl.hist(data,bins=30,density=True,histtype='stepfilled',
         color='steelblue',edgecolor='none',alpha=0.5)
pl.show()
```

```python
x1=np.random.normal(0,0.8,1000)
x2=np.random.normal(-2,1,1000)
x3=np.random.normal(3,2,1000)
```

```python
kwargs=dict(histtype='stepfilled',bins=40,alpha=0.5,density=True)
pl.hist(x1,**kwargs)
pl.hist(x2,**kwargs)
pl.hist(x3,**kwargs)
pl.show()
```

```python
pl.hist(x1,alpha=0.5,density=True,bins=40)
pl.hist(x2,alpha=0.5,density=True,bins=40)
pl.hist(x3,alpha=0.5,density=True,bins=40)
pl.show()
# its same as above kwargs function
# here we need to specify each and every time. there just once.
```

```python
import matplotlib.pyplot as pl
# pl.style.use('classic')
import numpy as np
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
rng=np.random.RandomState(0)
x=np.linspace(0,10,500)
y=np.cumsum(rng.randn(500,6),0)
pl.plot(x,y)
pl.legend('ABCDEF',ncol=2,loc='best')
pl.show()
```

```python
sns.set()
pl.plot(x,y)
pl.legend('ABCDEF',ncol=2,loc='best')
pl.show()
```

```python
data=np.random.multivariate_normal([0,0],[[5,2],[2,2]],size=2000)
data=pd.DataFrame(data,columns=['x','y'])
for col in 'xy':
    pl.hist(data[col],density=True,alpha=0.5)
```

```python
for col in 'xy':
    sns.kdeplot(data[col])
```

```python
pl.figure(figsize=(10,6))
sns.distplot(data['x'])
sns.distplot(data['y'])
pl.show()
```

```python
sns.kdeplot(data)
```

```python
sns.jointplot("x","y",data,s=5,kind='scatter')
pl.show()
```

```python
with sns.axes_style('white'):
    sns.jointplot("x","y",data,s=5,kind='kde')
```

```python
with sns.axes_style('white'):
    sns.jointplot('x','y',data,kind='hex')
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
tips['tip_pct'] = 100 * tips['tip'] / tips['total_bill']
grid=sns.FacetGrid(tips,row='sex',col='time')
grid.map(pl.hist,'tip_pct',bins=np.linspace(0,40,15))
```

```python
# Factor plots
with sns.axes_style(style='ticks'):
    g=sns.factorplot('day','total_bill','sex',data=tips,kind='box')
    g.set_axis_labels('Day','Total Bill')
```

```python
with sns.axes_style(style='ticks'):
    g=sns.factorplot('day','total_bill','sex',data=tips,kind='violin')
    g.set_axis_labels('Day','Total Bill')
```

```python
# Joint Distributions
with sns.axes_style('white'):
    sns.jointplot('total_bill','tip',data=tips,kind='hex')
```

```python
sns.jointplot('total_bill','tip',data=tips,kind='reg')
```

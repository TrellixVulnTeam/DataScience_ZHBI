```python
import pandas as pd
import numpy as np
```

```python
date=np.array(['5/1/2017','5/2/2017','5/3/2017','5/1/2017','5/2/2017',
     '5/3/2017','5/1/2017','5/2/2017','5/3/2017','5/1/2017'])
city=np.array(['NY','NY','NY','Bei','Bei','Bei','LA','LA','LA','DA'])
temp=np.random.randint(1,100,(10))
hum=np.random.randint(1,100,(10))
df=pd.DataFrame({"Date":date,
              "city":city,
              "Temperature":temp,
              "Humidity":hum})
df
```

```python
df.pivot(index="Date",columns="city")
# for pivot values need not be same
```

```python
df.pivot(index="city",columns="Date",values="Humidity")
```

```python
df.pivot(index="city",columns="Date",values="Temperature")
```

```python
df.pivot_table(index="Temperature",columns="Date")
```

```python
df.pivot_table(index="Humidity",columns="Date")
```

```python
date=np.array(['5/1/2017','5/1/2017','5/1/2017','5/2/2017','5/2/2017',
     '5/2/2017','5/3/2017','5/3/2017','5/3/2017','5/1/2017'])
city=np.array(['NY','NY','NY','Bei','Bei','Bei','LA','LA','LA','DA'])
temp=np.random.randint(1,100,(10))
hum=np.random.randint(1,100,(10))
df=pd.DataFrame({"Date":date,
              "city":city,
              "Temperature":temp,
              "Humidity":hum})
df
```

```python
df.pivot_table(index="city",columns="Date",aggfunc='sum')
# now this grouped cities on the basis of date and summed
# all the values which are similar to same group
```

```python
df.pivot_table(index="city",columns="Date",aggfunc='mean')
# now this grouped cities on the basis of date and mean is default
# all the values which are similar to same group
```

```python
df.pivot_table(index="city",columns="Date",margins=True)
```

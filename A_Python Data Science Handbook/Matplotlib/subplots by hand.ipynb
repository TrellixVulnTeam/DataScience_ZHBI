```python
import matplotlib.pyplot as pl
import numpy as np
import pandas as pd
```

```python
ax1=pl.axes()
ax2=pl.axes([0.65,0.65,0.2,0.2])
```

```python
fig=pl.figure()
ax1=fig.add_axes([0.1,0.5,0.8,0.4],
                xticklabels=[],ylim=(-1.2,1.2))
ax2=fig.add_axes([0.1,0.1,0.8,0.4],
                xticklabels=[],ylim=(-1.5,1.5))
x=np.linspace(0,10)
ax1.plot(np.sin(x))
ax2.plot(np.cos(x))
```

```python
import matplotlib.pyplot as pl
IPython_default=pl.rcParams.copy()
import warnings
warnings.filterwarnings("ignore")
import numpy as np
```

```python
pl.style.available[:5]
pl.style.use('dark_background')
```

```python
def hist_plot():
    np.random.seed(0)
    fig,ax=pl.subplots(1,2,figsize=(11,4))
    ax[0].hist(np.random.randn(1000))
    for i in range(3):
        ax[1].plot(np.random.rand(10))
    ax[1].legend(['a','b','c'],loc="upper right")
```

```python
pl.rcParams.update(IPython_default)
```

```python
hist_plot()
```

```python
with pl.style.context('fivethirtyeight'):
    hist_plot()
```

```python
with pl.style.context('ggplot'):
    hist_plot()
```

```python
with pl.style.context('bmh'):
    hist_plot()
```

```python
with pl.style.context('dark_background'):
    hist_plot()
```

```python
with pl.style.context('grayscale'):
    hist_plot()
```

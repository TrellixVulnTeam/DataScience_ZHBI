```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as pl
```

```python
df=pd.read_csv('untitled.txt',sep=',').drop(columns=['Unnamed: 0'])
```

```python
df.dropna(inplace=True)
```

```python
pl.figure(figsize=(10,7))
sns.countplot(df['Time since posted'])
pl.xticks(rotation=-60)
pl.show()
```

```python
df.info()
```

```python
pl.figure(figsize=(10,7))
sns.distplot(df['Likes'],bins=20)
pl.xticks(rotation=-60)
pl.show()
```

```python
hashtags=[]
for txt in df['Hashtags']:
    items=txt.split('#')
    items= [x for x in items if x !='']
    hashtags.extend(items)
hashtags[:10]
```

```python
captions=[]
for txt in df['Caption']:
    items=txt.split('#')
    items= [x for x in items if x !='']
    captions.extend(items)
captions[:10]
```

```python
captions=' '.join(captions)
captions[:100]
```

```python
hashtags=' '.join(hashtags)
hashtags[:100]
```

```python
from wordcloud import WordCloud, STOPWORDS
```

```python

wc=WordCloud(height=800,width=800,
            stopwords=STOPWORDS,
             background_color='white',
             min_font_size=10
            ).generate(hashtags)
pl.figure(figsize=(10,8),facecolor=None)
pl.title("Hashtag wordcloud")
pl.grid(False)
pl.imshow(wc)
pl.style.use('seaborn-whitegrid')
pl.rcParams.update({'font.size':25})
pl.axis('off')
pl.show()
```

```python
wc=WordCloud(height=800,width=800,
            stopwords=STOPWORDS,
             background_color='white',
             min_font_size=10
            ).generate(captions)
pl.figure(figsize=(10,8),facecolor=None)
pl.title("Caption wordcloud")
pl.grid(False)
pl.imshow(wc)
pl.style.use('seaborn-whitegrid')
pl.rcParams.update({'font.size':25})
pl.axis('off')
pl.show()
```

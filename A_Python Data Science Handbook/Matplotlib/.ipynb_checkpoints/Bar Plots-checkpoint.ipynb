```python
import seaborn as sns
planets=sns.load_dataset('planets')
planets.head()
```

```python
with sns.axes_style('white'):
    g=sns.factorplot("year",data=planets,kind='count',aspect=2)
    g.set_xticklabels(step=5)
#     here step means we are checking for 5,5 years data gap
```

```python
g=sns.factorplot('year',data=planets,kind='count',aspect=2,hue='method',order=range(2001,2008))
g.set_ylabels('Number of planets Discovered')
```

```python
g=sns.factorplot('year',data=planets,kind='count',aspect=2,hue='method',order=range(2008,2015))
g.set_ylabels('Number of planets Discovered')

```

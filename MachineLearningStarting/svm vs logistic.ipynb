```python
from sklearn.datasets import make_regression
X,y=make_regression(n_samples=1000,n_features=1,n_targets=1,noise=50)
```

```python
from sklearn.linear_model import Ridge
```

```python
X,y=make_regression(n_features=5,n_samples=1000)
```

```python
linear=LinearRegression().fit(X,y)
ridge=Ridge(alpha=2).fit(X,y)
lasso=Lasso(alpha=2).fit(X,y)
from sklearn.linear_model import ElasticNet
elast=ElasticNet().fit(X,y)
```

```python
fig,axes=pl.subplots(3,2,figsize=(15,15))
a=np.linspace(-3,3,1000)

axes[0][0].scatter(X[:,0],y)
axes[0][0].plot(a,linear.coef_[0]*a+linear.intercept_,c='r')
axes[0][0].plot(a,ridge.coef_[0]*a+ridge.intercept_,c='y')
axes[0][0].plot(a,lasso.coef_[0]*a+lasso.intercept_,c='c')


axes[0][1].scatter(X[:,1],y)
axes[0][1].plot(a,linear.coef_[1]*a+linear.intercept_,c='r')
axes[0][1].plot(a,ridge.coef_[0]*a+ridge.intercept_,c='y')
axes[0][1].plot(a,lasso.coef_[0]*a+lasso.intercept_,c='c')

axes[1][0].scatter(X[:,2],y)
axes[1][0].plot(a,linear.coef_[2]*a+linear.intercept_,c='r')
axes[1][0].plot(a,ridge.coef_[0]*a+ridge.intercept_,c='y')
axes[1][0].plot(a,lasso.coef_[0]*a+lasso.intercept_,c='c')

axes[1][1].scatter(X[:,3],y)
axes[1][1].plot(a,linear.coef_[3]*a+linear.intercept_,c='r')
axes[1][1].plot(a,ridge.coef_[0]*a+ridge.intercept_,c='y')
axes[1][1].plot(a,lasso.coef_[0]*a+lasso.intercept_,c='c')
                 
axes[2][0].scatter(X[:,4],y)
axes[2][0].plot(a,linear.coef_[4]*a+linear.intercept_,c='r')
axes[2][0].plot(a,ridge.coef_[0]*a+ridge.intercept_,c='y')
axes[2][0].plot(a,lasso.coef_[0]*a+lasso.intercept_,c='c')
```

```python
df=pd.DataFrame(columns=['Coefficient1','Coefficient2','Coefficient3','Coefficient4','Coefficient5'])
df
```

```python
for i in range(4):
    if i==0:
        df.loc[i]=[linear.coef_[j] for j in range(5)]
        
    if i==1:
        df.loc[i]=[ridge.coef_[j] for j in range(5)]

    if i==2:
        df.loc[i]=[lasso.coef_[j] for j in range(5)]
        
    if i==3:
        df.loc[i]=[elast.coef_[j] for j in range(5)]

df['Intercept']=[linear.intercept_,ridge.intercept_,lasso.intercept_,elast.intercept_]
```

```python
X,y=sklearn.datasets.make_blobs(n_samples=1000,centers=2,cluster_std=2)
svc=SVC()
svc.fit(X,y)
svc.decision_function
```

```python
mglearn.discrete_scatter(X[:,0],X[:,1],y)
```

```python
pl.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap=pl.cm.Accent)
ax = pl.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# create grid to evaluate model
xx = np.linspace(xlim[0], xlim[1],30)
yy = np.linspace(ylim[0], ylim[1], 30)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = svc.decision_function(xy).reshape(XX.shape)

# plot decision boundary and margins
ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
           linestyles=['--', '-', '--'])
# plot support vectors
ax.scatter(svc.support_vectors_[:, 0], svc.support_vectors_[:, 1], s=100,
           linewidth=1, facecolors='none', edgecolors='k')
pl.show()
```

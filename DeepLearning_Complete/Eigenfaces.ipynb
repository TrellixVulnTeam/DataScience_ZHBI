```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as pl
from PIL import Image
from skimage import color
```

```python
a=Image.open('risu.jpg')
```

```python
img=np.array(a)
```

```python
img=color.rgb2gray(img)
```

```python
img=Image.fromarray(img)
```

```python
pl.imshow(np.array(img))
```

```python
newsize=(1000,1024)
img=img.resize(newsize)
```

```python
img=np.array(img)
```

```python
img.shape
```

```python
img2=img.copy()
```

```python
pl.imshow(img2)
```

```python
pca=PCA(n_components=2).fit_transform(img2)
```

```python
pca[:,0].reshape(32,32)
```

```python
pl.plot(pca[:,0])
```

```python
pl.plot(pca[:,1])
```

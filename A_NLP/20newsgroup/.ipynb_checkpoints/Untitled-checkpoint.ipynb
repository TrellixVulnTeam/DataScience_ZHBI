```python
import nltk
from nltk.corpus import stopwords
from nltk.stem.wordnet import WordNetLemmatizer
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
lemitizer=WordNetLemmatizer()
```

```python
def cleanText(text):
    text=text.replace('\n','')
    text=text.replace('>','')
    text=text.replace('=','')
    text=text.replace("\"",'')
    text=text.replace('+','')
    text=text.replace('-','')

    
    words=text.split()
    words=[stemmer.stem(word.lower()) for word in words if word.lower() not in stopwords.words('english')]
    text=' '.join(words)
    return text
```

```python
atheism=cleanText(atheism)
graphics=cleanText(graphics)
windows_misc=cleanText(windows_misc)
hardware=cleanText(hardware)
mac=cleanText(mac)
windows=cleanText(windows)
forsale=cleanText(forsale)
autos=cleanText(autos)
motorcycles=cleanText(motorcycles)
baseball=cleanText(baseball)
hockey=cleanText(hockey)
crypt=cleanText(crypt)
electronics=cleanText(electronics)
med=cleanText(med)
space=cleanText(space)
christian=cleanText(christian)
guns=cleanText(guns)
mideast=cleanText(mideast)
politics=cleanText(politics)
religion=cleanText(religion)
```

```python
import matplotlib.pyplot as pl
import numpy as np
pl.style.use('seaborn-whitegrid')
```

```python
fig=pl.figure()
ax=pl.axes()
```

```python
x=np.linspace(0,10,1000)
ax.plot(x,np.sin(x))
```

```python
pl.plot(x,np.sin(x))
```

```python
pl.plot(x,np.sin(x),label='Sin wave')
pl.plot(x,np.cos(x),label='Cos wave')
pl.legend(loc="best")
```

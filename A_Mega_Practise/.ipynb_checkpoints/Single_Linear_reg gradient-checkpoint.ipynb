```python
from sklearn.datasets import make_regression 
import matplotlib.pyplot as pl 
import numpy as np 
import pandas as pd 
X,y=make_regression(n_samples=300,n_features=1,noise=10) 
pl.figure(figsize=(10,7)) 
y=y.reshape(-1,1)
print(y.shape)
pl.scatter(X,y,c=y) 
pl.show() 
```

```python
import numpy as np
m=0
c=0
n=len(X)

L=0.01
reps=1000

listoferror=[]

for i in range(reps):
    y_pred=m*X+c
    d_m=(-2/n)*np.sum(X*(y-y_pred), axis=0)
    d_c=(-2/n)*np.sum(y-y_pred)
    #print(d_m, d_c)
    m=m-L*d_m
    c=c-L*d_c
    
    errors=np.square(np.subtract(y,y_pred))
    err=np.mean(errors)
    listoferror.append(err)
print(m,c)
#print(listoferror)
```

```python
pl.figure(figsize=(10,7))
y_pred=m*X+c
pl.scatter(X,y,c=y)
pl.plot(X,y_pred,c="r")
pl.show()
```

```python
pl.plot(listoferror)
```

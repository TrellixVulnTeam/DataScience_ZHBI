```python
fig = pl.figure(figsize=(10,10))
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax1.hist(np.random.randn(100),bins=20,color='k',alpha=0.3)
ax2.scatter(np.arange(30),np.arange(30)+3*np.random.randn(30))
pl.plot(np.random.randn(50).cumsum(),'k--')
```

```python
from numpy.random import randn
pl.figure(figsize=(8,8))
pl.plot(randn(30).cumsum(),'ko:')
```

```python
pl.figure(figsize=(8,8))
pl.plot(randn(30).cumsum(),color='k',linestyle='dotted',marker='o')
```

```python
data=randn(30).cumsum()
pl.figure(figsize=(8,8))
pl.plot(data,'k--',label='Default')
pl.plot(data,'k-',drawstyle='steps-post',label='steps-post')
pl.legend(loc='best')
```

```python
fig=pl.figure(figsize=(10,6))
ax=fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum())
ticks=ax.set_xticks([0,250,500,750,1000])
labels=ax.set_xticklabels(['one','two','three','four','five'])
ax.set_title("My plot with ticks")
ax.set_xlabel('Stages')
```

```python
from numpy.random import randn
fig=pl.figure(figsize=(10,8));
ax=fig.add_subplot(1,1,1)
ax.plot(randn(1000).cumsum(),'k',label='one')
ax.plot(randn(1000).cumsum(),'k--',label='two')
ax.plot(randn(1000).cumsum(),'k.',label='three')
ax.legend()
```

```python
from datetime import datetime
fig=pl.figure(figsize=(10,8))
ax=fig.add_subplot(1,1,1)
data=pd.read_csv('dates.csv',names=['Spx'],index_col=0,parse_dates=True)
spx=data['Spx']
spx.plot()
crisis_data = [
(datetime(2007, 10, 11), 'Peak of bull market'),
(datetime(2008, 3, 12), 'Bear Stearns Fails'),
(datetime(2008, 9, 15), 'Lehman Bankruptcy')
]
for date,label in crisis_data:
    ax.annotate(label, xy=(date, spx.asof(date) + 50),
    xytext=(date, spx.asof(date) +200),
    arrowprops=dict(facecolor='black', headwidth=6, width=0.2,
    headlength=4),
    horizontalalignment='left', verticalalignment='top')

ax.set_xlim(['1/1/2007', '1/1/2011'])
ax.set_ylim([600, 1800])
ax.set_title('Important dates in the 2008-2009 financial crisis')


data.head()
```

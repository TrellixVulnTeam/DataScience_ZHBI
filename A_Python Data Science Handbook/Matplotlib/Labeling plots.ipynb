```python
import matplotlib.pyplot as pl
import numpy as np
pl.style.use('seaborn-whitegrid')
```

```python
x=np.linspace(0,10,100)
```

```python
pl.plot(x,np.sin(x))
pl.axis([0,10,-1.01,1.01])
pl.title("A Sine Curve")
pl.xlabel("X-Axis")
pl.ylabel("Y-Axis")
```

```python
pl.plot(x,np.sin(x),'--',color='g',label="Sine Wave")
pl.plot(x,np.cos(x),':',color='r',label="Cos Wave")
pl.axis('equal')
pl.legend()
```

```python
# 19mcmi23 Rishav Paudel

from sklearn.datasets import load_iris
import matplotlib.pyplot as pl
import numpy as np
import pandas as pd
```

```python
data=load_iris()
iris=data["data"]
```

```python
# Iris classes
type_setosa=[[0 for x in range(50)]for y in range(4)]
type_versicolor=[[0 for x in range(50)]for y in range(4)]
type_verginica=[[0 for x in range(50)]for y in range(4)]
```

```python
for i in range(0,50):
    type_setosa[0][i]=iris[i][0]
    type_setosa[1][i]=iris[i][1]
    type_setosa[2][i]=iris[i][2]
    type_setosa[3][i]=iris[i][3]

    type_versicolor[0][i]=iris[50+i][0]
    type_versicolor[1][i]=iris[50+i][1]
    type_versicolor[2][i]=iris[50+i][2]
    type_versicolor[3][i]=iris[50+i][3]
    
    type_verginica[0][i]=iris[100+i][0]
    type_verginica[1][i]=iris[100+i][1]
    type_verginica[2][i]=iris[100+i][2]
    type_verginica[3][i]=iris[100+i][3]
```

```python
# plotting histogram of four dimensions of original dataset
fig=pl.figure(7)
for i in range(0,4):
    pl.hist(type_setosa[i],color="blue",label="Setosa")
    pl.hist(type_versicolor[i],color="red",label="versicolor")    
    pl.hist(type_verginica[i],color="yellow",label="verginica")
    pl.title("Histogram of Dimension:{} of Original Iris".format(i))
    pl.legend()
    pl.show()
```

```python
# finding optimal w
mean_setosa=np.array([np.mean(type_setosa[0]),np.mean(type_setosa[1]),
                     np.mean(type_setosa[2]),np.mean(type_setosa[3])])
mean_versicolor=np.array([np.mean(type_versicolor[0]),np.mean(type_versicolor[1]),
                     np.mean(type_versicolor[2]),np.mean(type_versicolor[3])])
mean_verginica=np.array([np.mean(type_verginica[0]),np.mean(type_verginica[1]),
                     np.mean(type_verginica[2]),np.mean(type_verginica[3])])
mean_setosa,mean_versicolor,mean_verginica
```

```python
iris_mean=(mean_setosa+mean_verginica+mean_versicolor)/3
iris_mean
```

```python
# finding S_B
b_mean_setosa=mean_setosa-iris_mean
b_mean_versicolor=mean_versicolor-iris_mean
b_mean_verginica=mean_verginica-iris_mean

s_b_setosa=np.dot(b_mean_setosa[np.newaxis].T,b_mean_setosa[np.newaxis])
s_b_versicolor=np.dot(b_mean_versicolor[np.newaxis].T,b_mean_versicolor[np.newaxis])
s_b_verginica=np.dot(b_mean_verginica[np.newaxis].T,b_mean_verginica[np.newaxis])

s_b_iris=s_b_setosa+s_b_verginica+s_b_versicolor
s_b_iris
```

```python
setosa_sc_mat=np.zeros((4,4))
versicolor_sc_mat=np.zeros((4,4))
verginica_sc_mat=np.zeros((4,4))
for i in range(0,50):
#SETOSA
    setosa_row=np.array([type_setosa[0][i],type_setosa[1][i],
                        type_setosa[2][i],type_setosa[3][i]])
    setosa_row.reshape(4,1)
    mean_setosa.reshape(4,1)
    var_setosa=(setosa_row-mean_setosa)
    setosa_sc_mat+=np.dot(var_setosa[np.newaxis].T,var_setosa[np.newaxis])
#VERSICOLOR   
    versicolor_row=np.array([type_versicolor[0][i],type_versicolor[1][i],
                        type_versicolor[2][i],type_versicolor[3][i]])
    versicolor_row.reshape(4,1)
    mean_versicolor.reshape(4,1)
    var_versicolor=(versicolor_row-mean_versicolor)
    versicolor_sc_mat+=np.dot(var_versicolor[np.newaxis].T,var_versicolor[np.newaxis])
#VERGINICA    
    verginica_row=np.array([type_verginica[0][i],type_verginica[1][i],
                        type_verginica[2][i],type_verginica[3][i]])
    verginica_row.reshape(4,1)
    mean_verginica.reshape(4,1)
    var_verginica=(verginica_row-mean_verginica)
    verginica_sc_mat+=np.dot(var_verginica[np.newaxis].T,var_verginica[np.newaxis])
    
```

```python
s_w_iris=setosa_sc_mat+verginica_sc_mat+verginica_row
eig_vals_iris,eig_vec_iris=np.linalg.eig(np.dot(np.linalg.inv(s_w_iris),s_b_iris))
eig_pairs_iris=[(np.abs(eig_vals_iris[i]),eig_vec_iris[:,i])for i in range(len(eig_vals_iris))]
eig_pairs_iris=sorted(eig_pairs_iris,key=lambda k: k[0],reverse=True)
print("Eigen values by sorted:\n")
for i in eig_pairs_iris:
    print(i)
```

```python
optimal_w_iris=np.hstack((eig_pairs_iris[0][1].reshape(4,1),eig_pairs_iris[1][1].reshape(4,1)))
print("Optimal w_iris:\n",optimal_w_iris)
```

```python
y_setosa_opt=np.dot(np.transpose(type_setosa),optimal_w_iris)
y_versicolor_opt=np.dot(np.transpose(type_versicolor),optimal_w_iris)
y_verginica_opt=np.dot(np.transpose(type_verginica),optimal_w_iris)
```

```python
def plot_step_LDA():
    ax=pl.subplot(111)
    pl.scatter(y_setosa_opt[:,0],
              y_setosa_opt[:,1],
              marker="^",
              color="blue",
              alpha=0.5,
              label="Setosa")
    pl.scatter(y_verginica_opt[:,0],
              y_verginica_opt[:,1],
              marker="s",
              color="red",
              alpha=0.5,
              label="Virginica")
    pl.scatter(y_versicolor_opt[:,0],
              y_versicolor_opt[:,1],
              marker="o",
              color="green",
              alpha=0.5,
              label="Setosa")
    pl.xlabel('LD1')
    pl.ylabel('LD2')
    leg=pl.legend(loc="upper right",fancybox=True)
    leg.get_frame().set_alpha(0.5)
    pl.title("LDA: Iris projection into first 2 linear Discriminants")
    pl.tick_params(axis="both",which="both",bottom="off",top="off",
                  labelbottom="on",left="off",right="off",labelleft="on")
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.spines["bottom"].set_visible(False)
    ax.spines["left"].set_visible(False)
    pl.grid()
    pl.tight_layout
    pl.show()
plot_step_LDA()
```

```python
# creating histograms for the reduced LDA in dimension1
y_setosa_opt_1=np.dot(eig_pairs_iris[0][1],type_setosa)
y_verginica_opt_1=np.dot(eig_pairs_iris[0][1],type_verginica)
y_versicolor_opt_1=np.dot(eig_pairs_iris[0][1],type_versicolor)
pl.hist(y_setosa_opt_1,label="y_setosa")
pl.hist(y_verginica_opt_1,label="y_virginica")
pl.hist(y_versicolor_opt_1,label="y_versicolor")
pl.legend()
pl.show()
# Conclusion: We can see how well they are separated
```

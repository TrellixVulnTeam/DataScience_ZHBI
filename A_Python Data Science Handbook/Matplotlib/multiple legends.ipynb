```python
import numpy as np
import matplotlib.pyplot as pl
from matplotlib.legend import Legend
fig,axes=pl.subplots()
lines=[]
styles=['-','--','-.',':']
x=np.linspace(0,10,1000)
for i in range(4):
    lines+=axes.plot(x,np.sin(x-i*np.pi/2),styles[i],color='black')
axes.axis('equal')
axes.legend(lines[:2],['Line A','line B'],loc='upper right',frameon=False)
leg=Legend(axes,lines[2:],['Line C','line D'],loc='lower right',frameon=False)
axes.add_artist(leg)
```

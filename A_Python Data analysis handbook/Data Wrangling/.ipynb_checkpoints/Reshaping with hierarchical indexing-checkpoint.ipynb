```python
import numpy as np
import pandas as pd
```

```python
data = pd.DataFrame(np.arange(6).reshape((2, 3)),
        index=pd.Index(['Ohio', 'Colorado'], name='state'),
        columns=pd.Index(['one', 'two', 'three'],
        name='number'))
data
```

```python
print(data.stack(),"\n\n")
print(data.unstack())
```

```python
result=data.stack()
result.unstack(0)
```

```python
result.unstack('state')
```

```python
df = pd.DataFrame({'key': ['foo', 'bar', 'baz'],
                    'A': [1, 2, 3],
                    'B': [4, 5, 6],
                    'C': [7, 8, 9]})
df
```

```python
melted=pd.melt(df,['key'])
melted
```

```python
melted.pivot('key','variable')
# melt and pivot are inversly proportional to each other :)
```

```python
from itertools import permutations
a=permutations(['a','b','c'])
b=[]
for i in a:
    b.extend(i)
print(b)
```

```python
df=pd.DataFrame({'letter':b,
                 'value':np.random.randint(1,10,18),
                 'Sno':np.arange(18)}).set_index('Sno')
df.columns.name="S"
```

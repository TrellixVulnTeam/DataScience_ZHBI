```python
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(1,11,1)
y=np.array([chr(i) for i in range(65,75)])

plt.figure(figsize=(10,5))
plt.title("Vertical bar graph")
plt.xlabel("Fruits name")
plt.ylabel("Quantity")
plt.bar(y,x,color="#808080")
```

```python
plt.figure(figsize=(10,7))

plt.barh(y,x,color="r")
plt.title("Horizontal bar graph")
plt.xlabel("Quantity")
plt.ylabel("Fruits name")
```

```python
a={"Apple":100,"mango":120,"Orange":50,"Grape":80}
name=list(a.keys())
value=list(a.values())
name,value
plt.figure(figsize=(10,5))
plt.bar(name,value,color="m")
plt.grid(True)
plt.title("Horizontal bar graph")
plt.xlabel("Fruits name")
plt.ylabel("Quantity")
```

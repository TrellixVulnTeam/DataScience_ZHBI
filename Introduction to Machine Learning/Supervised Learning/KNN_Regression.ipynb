```python
from sklearn.neighbors import KNeighborsRegressor
import mglearn
from sklearn.model_selection import train_test_split
import pandas as pd
X,y=mglearn.datasets.make_wave(n_samples=20)
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0)
reg=KNeighborsRegressor(n_neighbors=3)
reg.fit(X_train,y_train)
```

```python
a=reg.predict(X_test)
a
```

```python
pd.DataFrame({"Actual":y_test,"Predicted":a})
```

```python
test=[]
train=[]
for i in range(1,10):
    reg=KNeighborsRegressor(n_neighbors=i).fit(X_train,y_train)
    test.append(reg.score(X_test,y_test))
    train.append(reg.score(X_train,y_train))
```

```python
reg.score(X_train,y_train)
```

```python
import matplotlib.pyplot as pl
```

```python
pl.plot(train,linestyle=':',label="Training")
pl.plot(test,label="Testing")
pl.legend()
```

```python
import numpy as np
fig,axes=pl.subplots(1,3,figsize=(15,4))
line=np.linspace(-3,3,1000).reshape(-1,1)
for n,ax in zip([1,3,9],axes):
    reg=KNeighborsRegressor(n_neighbors=n)
    reg.fit(X_train,y_train)
    ax.plot(line,reg.predict(line))
    ax.plot(X_train,y_train,'^',c=mglearn.cm2(0),markersize=8)
    ax.plot(X_test,y_test,'^',c=mglearn.cm2(1),markersize=8)
    ax.set_title("{}neighbor(s)\ntrain score:{:2f} test score:{:2f}".format(
                n,reg.score(X_train,y_train),
                reg.score(X_test,y_test)))    
    ax.set_xlabel("Feature")
    ax.set_ylabel("Target")
    axes[0].legend(["Model predictions", "Training data/target",
"Test data/target"], loc="best")
```

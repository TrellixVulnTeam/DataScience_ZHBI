```python
import matplotlib.pyplot as pl
import numpy as np
pl.style.use('seaborn-whitegrid')
```

```python
fig=pl.figure()
ax=pl.axes()
```

```python
x=np.linspace(0,10,1000)
ax.plot(x,np.sin(x))
```

```python
pl.plot(x,np.sin(x))
```

```python
pl.plot(x,np.sin(x),label='Sin wave')
pl.plot(x,np.cos(x),label='Cos wave')
pl.legend(loc="best")
```

```python
pl.figure(figsize=(10,6))
pl.plot(x, np.sin(x - 0), color='blue') # specify color by name
pl.plot(x, np.sin(x - 1), color='g') # short color code (rgbcmyk)
pl.plot(x, np.sin(x - 2), color='0.75') # Grayscale between 0 and 1
pl.plot(x, np.sin(x - 3), color='#FFDD44') # Hex code (RRGGBB from 00 to FF)
pl.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 and 1
pl.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
```

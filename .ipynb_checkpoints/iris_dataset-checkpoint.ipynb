```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as pl
import seaborn as sns
import numpy as np
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv("iris.csv")
data=data.rename(columns={"SepalLengthCm":"sepal_len",
                          "SepalWidthCm":"sepal_wid",
                          "PetalLengthCm":"petal_len",
                          "PetalWidthCm":"petal_wid"})
print(data.info())
data.head()
```

```python
x=data[["petal_len","petal_wid","sepal_len","sepal_wid"]]
y=data[["Species"]]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
x_train.size,x_test.size,y_train.size,y_test.size
reg=LogisticRegression()
reg.fit(x_train,y_train)
```

```python
preds=reg.predict(x_test)
pp=pd.DataFrame({"Predicted":preds})
df=pd.concat([x,pp],join="inner",axis=1)
df
```

```python
confusion_matrix(y_test,preds)
```

```python
accuracy_score(y_test,preds)
```

```python
ranf=RandomForestClassifier()
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
ranf.fit(x_train,y_train)
```

```python
pred=ranf.predict(x_test)
pp=pd.DataFrame({"Predicted":pred})
df=pd.concat([x,pp],join="inner",axis=1)
df
```

```python
confusion_matrix(y_test,pred)
```

```python
accuracy_score(y_test,pred)
```

```python
np.equal(preds,pred)
```

```python
dect=DecisionTreeClassifier()
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
dect.fit(x_train,y_train)
```

```python
dectpr=dect.predict(x_test)
pp=pd.DataFrame({"Predicted":dectpr})
df=pd.concat([x,pp],join="inner",axis=1)
df
```

```python
confusion_matrix(y_test,dectpr)
```

```python
accuracy_score(y_test,dectpr)
```
